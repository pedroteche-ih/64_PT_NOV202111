## What are Numpy Arrays?

Numpy arrays are very close to **lists** in python. The main differences for us in Data Analysis are:

- There are *LOTS* of new mathematical methods that you can use in numpy arrays;
- Operations are much faster (if you are dealing with big data);
- Less Memory.

### 6. The `axis` concept - very useful

- The `axis` concept: 
    - axis = 0, first axis == `rows`
        - When we set axis=0, the function will **collapses** the rows and calculates the function
        
    - axis = 0, first axis == `columns`
        - When we set axis=1, the function will **collapses** the columns and calculates the function

### 7. Operations between `np.arrays`

```python
matriz_3_3 = np.array([[1, 1, 1],
                       [2, 2, 2],
                       [1, 2, 3]])
vetor_3 = np.array([5, 12, 1.8])
print(matriz_3_3)
print(vetor_3)
```

```python
vetor_3 + matriz_3_3
```

```python
matriz_6_3 = np.array([[1, 1, 1],
                       [2, 2, 2],
                       [1, 2, 3],
                       [1, 2, 3],
                       [1, 2, 3],
                       [1, 2, 3]])
```

```python
vetor_3 + matriz_6_3
```

```python
matriz_3_6 = np.array([[1, 1, 1, 2, 2, 2],
                       [1, 2, 3, 1, 2, 3],
                       [1, 2, 3, 1, 2, 3]])
print(matriz_3_6)
```

```python
vetor_3 + matriz_3_6
```

```python
ma_2_2 = np.full((2,2), 2)
ma_2_3 = np.full((2,2), 6)
```

```python
print(ma_2_2)
print(ma_2_3)
print(ma_2_2 + ma_2_3)
```

```python
print(matriz_3_3)
print(vetor_3)
print(vetor_3 * matriz_3_3)
```

```python
vetor_4 = np.array([1,2,3,4])
print(vetor_4)
```

```python
vetor_4 * matriz_3_3
```
